```python
#抓內容頁html檔
import requests
from bs4 import BeautifulSoup
import time

with open('mobile01-Taoyuan.txt', 'r', encoding='UTF-8') as f:
    #讀取內容頁網址
    info = f.read()
info.split('\n')
```

```python
for i in range(1,len(info.split('\n'))+1):
    if i%3 == 0:
        res = requests.get(info.split('\n')[i])
        #soup = BeautifulSoup(res.text, 'lxml')
        with open('D:\Project_Image\mobile01\桃園\{}.html'.format(i//3+1), 'w', encoding='UTF-8') as f:            
            f.write(res.text)
            time.sleep(3)
```
